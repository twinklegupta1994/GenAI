# List Comprehension :

#### we are going to use looping statements inside the list so that we can store the values inside the list in an iterable format and in list comprehention conditional statements are allowed as well.

### Syntax for list comprehention is :

### [expression for item in iterable if condition]

### expression --> what we want to print
### for items in iterable --> Looping statement
### if condition ---> conditional statement

```python
# to create a list that is storing the values from 0 to 100
list_items = [x for x in range(1,101)]
print(list_items)
```

```python
list_squared = [x**2 for x in range(1,26)]
print(list_squared)
```

```python
even_squared = [x**2 for x in range(1,26) if x %2 == 0]
print(even_squared)
```

### Nested List comprehention

#### we will be having the combination of multiple lists in our single list itself.

```python
pair = [[i, j] for i in list_squared for j in list_items]
print(pair)
```

```python
list1 = [1,2,3]
lsit2 = [4,5,6]
paried_list = [[i,j] for i in list1 for j in lsit2]
print(paried_list)
```
