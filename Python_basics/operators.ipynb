# Operators

## Operators helps to applying operation on the given operands
## Types of Operators :
1. Arithmetic Operators :

Addition(+)

Sunstraction (-)

Multiplication or Product (*)

Division (/)

Modulus (%)

Floor (//)

Exponent (a**b)

2. Comparition Opertors:

Greater that (>)

Less Than (<)

Greater than equal to (>=)

Less than equal to (<=)

Equality Check (==)

Not Equal to (!=)

3. Logical Operators:

and ---> Whether if both condition are true or not.

or ---> Whether if any of the operation is true or not

not ---> Negation or to inverse the result of the given operation.

```python
a=18
b=5
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b)
print(a//b)
print(a**b)
```
